```python
#### 
#import
import numpy as np
#lib affichage
import matplotlib.pyplot as plt
#lib traitement image
from PIL import ImageOps
from PIL import ImageFilter
from PIL import Image as pil
from PIL import ImageEnhance
```

```python
#lecture image
from pathlib import Path 

src = Path('C:/Users/Marc/Desktop/Projet_Traitement_image/ENID_v1.0_tracked/ENID_v1.0_tracked/frames/') ## CHEMIN  VERS fichier des images
listFiles = [f.resolve() for f in src.glob("*.jpg")]

len(listFiles)
image_original = pil.open(listFiles[40]) # lecture de l'image 0 Ã  12532
```

```python
image_original #************affichage de l'image original**************************
```

```python
image_contour_accentuation = image_original.filter(ImageFilter.EDGE_ENHANCE) #filtrage
image_contour_accentuation #*********************affichage image avec contour accentuÃ©****************
```

```python
# Traitement Histogramme
```

```python
img = np.array(image_contour_accentuation)
# GÃ©nÃ©rer et afficher l'histogramme
# Pour le normaliser : argument density=True dans plt.hist
# Pour avoir l'histogramme cumulÃ© : argument cumulative=True
n, bins, patches = plt.hist(img.flatten(), bins=range(256))
plt.show()
```

```python
im_hist_equalize = ImageOps.equalize(image_contour_accentuation, mask = None)
img = np.array(im_hist_equalize)
n, bins, patches = plt.hist(img.flatten(), bins=range(256))
plt.show()
```
